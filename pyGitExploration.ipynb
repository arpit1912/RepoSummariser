```python
def repo_details(owner,repo,token):
    query_url = f"https://api.github.com/repos/{owner}/{repo}"
    #print("Inside the repo Detail section")
    #print(query_url)
    params = {
        "state": "open",
    }
    headers = {'Authorization': f'token {token}'}
    r = requests.get(query_url, headers=headers, params=params)
    return r.json()
```

```python
def get_contributions_list(owner,repo,token):
    query_url = f"https://api.github.com/repos/{owner}/{repo}/contributors?page=1&per_page=10000"
    params = {
        "state": "open",
    }
    #print(query_url)
    headers = {'Authorization': f'token {token}'}
    r = requests.get(query_url, headers=headers, params=params)
    data = r.json()
    #if data.get('Link') != None:
        #pprint (r.headers)
    #print (data)
    return r.json()
```

```python
def calculated_contribution(owner,original_repo_contribution_data):
    total_contributions = 0
    user_contribution = 0
    for contributor in original_repo_contribution_data:
        if(contributor['contributions'] > 0):
            total_contributions += contributor['contributions']
            if(contributor['login'] == owner):
                print(contributor['login'], " this is the match we found")
                user_contribution = contributor['contributions']
        
            
    
    return total_contributions, user_contribution
```

```python
# Filter the Repos Based on the Number of Contributors and Find the Parent Repo name as well if exist

def filter_valid_repos(nonfiltered_repo_list):
    filtered_repo_list = []

    for user_repos in nonfiltered_repo_list:
        temp_repo_list = []
        for repo in user_repos:
            #print(repo['full_name'])
            if repo['fork'] is True:
                #print("in")
                owner,reponame = repo['full_name'].split("/")
                original_repo_data = repo_details(owner,reponame,token)
                if original_repo_data.get('source') != None:
                    print(original_repo_data['source']['full_name'])
                    main_owner,main_repo = original_repo_data['source']['full_name'].split("/")
                    original_repo_contribution_data = get_contributions_list(main_owner,main_repo,token)
                    total, user = calculated_contribution(owner,original_repo_contribution_data)
                    if(user > 0):
                        temp_repo_list.append(repo)
                        print(total, user)
                #pprint(original_repo_data)
                # do something
            else:
                print(repo['contributors_url'])

        filtered_repo_list.append(temp_repo_list)

    return filtered_repo_list
```

```python
#get all repositories of contributors
#NOTE : need to filter this list to get valid repos

def get_all_user_repos(user_names):
    unfiltered_repo_list = []
    for username in user_names:
        print(username)
        query_url = f"https://api.github.com/users/{username}/repos"
        pprint(query_url)
        params = {
            "state": "open",
        }
        headers = {'Authorization': f'token {token}'}
        r = requests.get(query_url, headers=headers, params=params)
        unfiltered_repo_list.append(r.json())
    
    return unfiltered_repo_list
```

```python
#get usernames of all contributors in a repo

def get_contributor_login(contributor_list):
    user_names = []
    for user in contributor_list:
        if user.get('login') != None:
            user_names.append(user['login'])
        #pprint(user)

    return user_names
```
